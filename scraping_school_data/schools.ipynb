```python
from bs4 import BeautifulSoup
import urllib.request as urllib
```

```python
school_url = 'https://www.seattleschools.org/directory/schools'
page = urllib.urlopen(school_url)
soup = BeautifulSoup(page, "html.parser")
```

```python
text = soup.find_all('li')
text = str(text)
text = text.split('<li>')
text
```

```python
school_list = []
school_urls = []


for item in text:
    if item[0:18] == '<a href="/schools/':
        for  i in range(len(item)):
            if item[i:(i+3)] == '" t':
                end = i
            
            
        school_urls.append("https://www.seattleschools.org"+item[9:end])  
        school_list.append(item[18:])
        
        
for url in school_urls:
    print(url)
    
school_list
```

```python
streets = []
zip_codes = []

j = 0
for school_url in school_urls:
    page = urllib.urlopen(school_url)
    soup = BeautifulSoup(page, "html.parser")
    address = soup.find('p')
    address = str(address)
    
    try:
            start = 0
            while address[start] != ">":
                start += 1

            end = start + 1
            while address[end] != "<":
                end += 1
                
            if address[start+1].isdigit() == True:    
                street = address[(start+1):end]
                streets.append(street)
                
                i = end
                while address[i].isdigit() == False:
                    i += 1
                zip_code = address[i:(i+5)]
                zip_codes.append(zip_code)
            else:
                street = 'unknown street'
                zip_code = 'unknown zip'
                streets.append(street)
                zip_codes.append(zip_code)
    except:
        street = 'unknown street'
        zip_code = 'unknown zip'
        streets.append(street)
        zip_codes.append(zip_code)
    
    print(j, " -- ", street,",",zip_code)
    j += 1
  
```

```python
school_names = []
school_types = []

checked_school_list = []

for school in school_list:
    #schools types
    checked_school_list.append(school)
    
    if school[0] == "h":
        school_types.append('high school')
    elif school[0] == "m":
        school_types.append('middle school')
    elif school[0] == "k":
        school_types.append('k-8')
    elif school[0] == "e":
        school_types.append('elementary school')
    elif school[0] == "o":
        school_types.append('option')
    elif school[0] == "a":
        school_types.append('alternative and service')

        
    for i in range(len(school)):
        # school name
        if school[(i-2):i] == '">':
            start = i
        if school[i:(i+3)] == "</a":
            end = i
   
    school_name = school[start:end]
    school_names.append(school_name)
    
```

```python
import pandas as pd

schoolDataSet = list(zip(school_names, school_types, streets, zip_codes))
df = pd.DataFrame(data = schoolDataSet, columns=['School','Type','Street Address','Zip Code'])
df
```

```python
df.to_csv("schools")
```
