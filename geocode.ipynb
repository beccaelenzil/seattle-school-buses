```python
import pandas as pd
import numpy as np

# read in schools table
df = pd.read_csv('schools.csv',encoding='latin1')
total_rows = int(df.size/4)
print(total_rows)
df
```

```python
# add option/alt column to schools data table
# remove rows that repeat school names

import time

option_alt = []

option_index = df['Type'] == 'option'
option_schools_table = df[option_index]
option_school_names = list(option_schools_table['School'])

alt_index = df['Type'] == 'alternative and service'
alt_schools_table = df[alt_index]
alt_school_names = list(alt_schools_table['School'])

all_school_df = df[0:105]
all_school_names = list(all_school_df['School'])
all_school_df

option_alt_schools = df[105:129]

for index, row in all_school_df.iterrows():
    if row['School'] in option_school_names:
        option_alt.append('option')
    elif row['School'] in alt_school_names:
        option_alt.append('alternative and service')
    else:
        option_alt.append('N/A')
        
all_school_df['option_alt'] = option_alt

total_rows = int(df.size/5)
i = 1
for index, row in option_alt_schools.iterrows():
    if row['School'] not in all_school_names:
        new_data = [row['School'],row['Type'],row['Street Address'],row['Zip Code'],row['Type']]
        all_school_df.loc[total_rows+i] = new_data
        i += 1
        #new_def = pd.DataFrame([row['School'],row['Type'],row['Street Address'],row['Zip Code'],row['Type']],
        #     columns=['School','Type','Street Address', 'Zip Code','option_alt'])
        #print(new_df)

#all_school_df = all_school_df.append()


all_school_df
```

```python
from opencage.geocoder import OpenCageGeocode

key = '63a88c0af7c646888e45bf4386df46d1'
geocoder = OpenCageGeocode(key)

long = []
lat = []

for index, row in all_school_df.iterrows():
        query = row["Street Address"]+", "+ row["Zip Code"]
        results = geocoder.geocode(query)
        
        if len(results) > 0:
            the_lat = results[0]['geometry']['lat']
            the_long = results[0]['geometry']['lng']

            lat.append(the_lat)
            long.append(the_long)
        else:
            the_lat = 'unknown'
            the_long = 'unknown'
            lat.append('unknown')
            long.append('unknown')
        
        print(the_lat, the_long)
                                                
```

```python
all_school_df['latitude'] = lat
all_school_df['longitude'] = long
all_school_df
```

```python
import matplotlib.pyplot as plt

lat_list = []
long_list = []
name_list = []

n = list(all_school_df['School'])

for i in range(len(lat)):
    if type(lat[i]) == float and lat[i] > 47:
        lat_list.append(lat[i])
        long_list.append(long[i])
        name_list.append(n[i])

        
#fig, ax = plt.subplots()
ax = plt.axes([0, 0, 1.5, 3])

for i,txt in enumerate(name_list):
    ax.annotate(txt, (lat_list[i], long_list[i]))

plt.scatter(lat_list,long_list, norm = True)
plt.show()
```
