```python
import pandas as pd
from datascience import *
import ast

#school_df = pd.read_pickle('cleaned_school_list.pk1')
school_df = Table.read_table('cleaned_school_list.csv')
school_df

cleaned_schools = []
geocode_dict = []
school_type = school_df.column('Type')
i = 0
for school in school_df.column('School'):
    if school[len(school)-1] == ' ':
        cleaned_schools.append(school[0:len(school)-1])
        print(school[0:len(school)-1])
    elif school == 'Cleveland':
        cleaned_schools.append('STEM')
    elif school == 'Aki  Kurose':    
        cleaned_schools.append('Aki Kurose')
    elif school == 'West Seattle' and school_type[i] == 'elementary school':
        cleaned_schools.append('West Seattle Elementary')
    elif school == 'West Seattle' and school_type[i] == 'high school':
        cleaned_schools.append('West Seattle High')
    else:
        cleaned_schools.append(school)
    i+=1
    
for geocode1 in school_df.column('geocode'):
    #print(geocode1)
    geocode_dict.append(ast.literal_eval(geocode1))
         
school_df.drop('School')
school_df = school_df.with_column('School', cleaned_schools)

school_df.drop('geocode')
school_df = school_df.with_column('geocode', geocode_dict)

school_df


school_df.to_csv('current_school_list.csv')
current_school_df = pd.read_csv('current_school_list.csv')
current_school_df.to_pickle('current_school_list.pk1')
current_school_df

current_school_df.get('geocode')
    
```
