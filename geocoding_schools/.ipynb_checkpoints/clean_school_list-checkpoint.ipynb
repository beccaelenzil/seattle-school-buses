```python
import pandas as pd
import numpy as np
from datascience import *

# read in schools table
school_df = pd.read_pickle('school_list.pk1')
school_df

# data science method
#school_df = Table.read_table('../school_list.csv')
#school_df
```

```python
school_names = school_df.get("School")
school_names

#print(school_names)

names = []

for name in school_names:
    k = name.find(",")
    if name == "King, Martin Luther, Jr.":
        new_name = "Martin Luther King, Jr."
        names.append(new_name)
    elif k != -1:
        new_name = name[(k+2):]+" "+name[0:k]
        names.append(new_name)
    else:
        names.append(name)
        
#print(names)

school_df['School'] = names
school_df

school_df.to_pickle('cleaned_school_list.pk1')
school_df

    
```
